**Calculation of cost function**

```python
def eval_mse(y, yhat):
    m = len(y)
    err = 0.0
    for i in range(m):
        err+=np.square(yhat[i]-y[i])    
    err = err / (2*m)
    return(err)
y_hat = np.array([2.4, 4.2])
y_tmp = np.array([2.3, 4.1])
eval_mse(y_hat, y_tmp)
```
